# Binning

```python
bins = [0,100000,300000, float('inf')]
labels = ["Low","Med","High"]

df["Income Band"] = pd.cut(df['estimated_income'], bins = bins, labels = labels, right = False)
df["Income Band"].value_counts().plot(kind = 'bar') #Counts how many entries fall into each income band.
plt.show()
```

```python
Categorical_cols = df[[
    "nationality", "occupation", "fee_structure", "loyality_classification",
    "amount_of_credit_cards", "properties_owned", "rish_weighting",
    "brid", "gender_id", "iaid"
]]

for col in Categorical_cols:
    print(f"\nValue counts for '{col}':")
    display(
        df[col].value_counts().reset_index()
    )
```

# univariate analysis

```python
for i, predictor in enumerate(df[["nationality", "occupation", "fee_structure", "loyality_classification","amount_of_credit_cards", "properties_owned", "rish_weighting","brid", "gender_id", "iaid","Income Band"]]):
    plt.figure(i) #- This initializes a new figure for each plot, so each countplot appears separately.
    sns.countplot(data = df, x=predictor)
```

# Bivariate analysis

```python
for i, predictor in enumerate(df[["nationality", "occupation", "fee_structure", "loyality_classification","amount_of_credit_cards", "properties_owned", "rish_weighting","brid", "gender_id", "iaid","Income Band"]]):
    plt.figure(i) #- This initializes a new figure for each plot, so each countplot appears separately.
    sns.countplot(data = df, x=predictor, hue = "nationality")
```

```python
for col in Categorical_cols:
    if col == "occupation":
        continue
    plt.figure(figsize = (8,4))
    sns.histplot(df[col])
    plt.title("Histogram of {} count".format(col))
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

# numerical Analysis

```python
numerical_cols = df[["estimated_income","superannuation_savings","credit_card_balance","bank_loans","bank_deposits","checking_accounts","saving_accounts","foreign_currency_account","business_lending"]].columns

#univariate analysis and visualization
plt.figure(figsize = (15,10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col], kde = True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

# HeatMaps

```python
numerical_cols = df[["estimated_income","superannuation_savings","credit_card_balance","bank_loans","bank_deposits","checking_accounts","saving_accounts","foreign_currency_account","business_lending"]].columns

Correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize = (12,12))
sns.heatmap(Correlation_matrix, annot=True, cmap = 'crest', fmt=".2f")
plt.title("correlation matrix")
plt.show()
```
